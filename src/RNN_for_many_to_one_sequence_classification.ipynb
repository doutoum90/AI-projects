```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import Input, Activation, Dense, Flatten, Dropout, LSTM
from tensorflow.keras.models import Model
```

```python
X1 = list()
X2 = list()

X = list()
y = list()

X1 = [x+4 for x in range(0, 160,4)]
X2 = [x+5 for x in range(0, 200, 5)]

X = np.column_stack((X1, X2))
# print(X)
# print(y)
X = np.array(X).reshape(20, 2, 2)
# print(X)
y = np.array([sum(i) for i in [sum(x) for x in X]])
```

```python
def create_lstm_one_to_one_model():
    input_layer = Input(shape=(2,2))
    lstm1 = LSTM(100, activation="relu", return_sequences=True)(input_layer)
    lstm2 = LSTM(50, activation="relu", return_sequences=True)(lstm1)
    lstm3 = LSTM(25, activation="relu")(lstm2)
    dense1 = Dense(10, activation="relu")(lstm3)
    output_layer = Dense(1)(dense1)
    model = Model(input_layer, output_layer)
    model.compile(loss="mse", optimizer='adam')
    return model
```

```python
from tensorflow.keras.utils import plot_model
model = create_lstm_one_to_one_model()
plot_model(model, to_file="../images/model_lstm_plot2.png", show_shapes=True, show_layer_names=True)
```

```python
model.summary()
model_history = model.fit(X, y, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt 
plt.plot(model_history.history['loss'], label="loss")
plt.legend(['train'], loc= "upper right")
```

```python
X_test = np.array([[200, 225],[204, 230]] ).reshape((1,2,2))
y_pred = model.predict(X_test, verbose=0)
```
