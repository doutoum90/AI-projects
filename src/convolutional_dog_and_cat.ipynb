```python
import numpy as np
import glob
from PIL import Image
import matplotlib.pyplot as plt
import os
import random
```

# Lecture de la dataset

```python
targets = []
features = []

files = glob.glob("../data/training_set/*.jpg")
random.shuffle(files)
for file in files:
    features.append(np.array(Image.open(file).resize((75,75))))
    target = [1, 0] if 'cat' in file else [0, 1]
    targets.append(target)
    
    
features = np.array(features)
targets = np.array(targets)

print(targets.shape, targets.shape)
```

# Afficher Quelques images

```python
from random import randint

for a in [randint(0, len(targets)) for _ in range(10)]:
    plt.imshow(features[a])
    plt.show()
```
