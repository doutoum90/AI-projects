```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import Input, Activation, Dense, Flatten, Dropout, LSTM
from tensorflow.keras.models import Model
```

```python
X1 = list()
X2 = list()

X = list()
y = list()

X1 = [(x+1)*4 for x in range(30)]
X2 = [(x+1)*5 for x in range(30)]

for x1, x2 in zip(X1, X2):
    output=list()
    output.append(x1-2)
    output.append(x2-3)
    y.append(output)

X = np.column_stack((X1, X2))
X = np.array(X).reshape(30, 1, 2)
y = np.array(y)
```

```python
def create_lstm_one_to_many_model():
    input_layer = Input(shape=(1,2))
    lstm1 = LSTM(100, activation="relu", return_sequences=True)(input_layer)
    lstm2 = LSTM(50, activation="relu", return_sequences=True)(lstm1)
    lstm3 = LSTM(25, activation="relu")(lstm2)
    dense1 = Dense(10, activation="relu")(lstm3)
    output_layer = Dense(2)(dense1)
    model = Model(input_layer, output_layer)
    model.compile(loss="mse", optimizer='adam')
    return model
```

```python
from tensorflow.keras.utils import plot_model
model = create_lstm_one_to_many_model()
plot_model(model, to_file="../images/model_lstm_plot3.png", show_shapes=True, show_layer_names=True)
```

```python
model_history = model.fit(X, y, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt 
plt.plot(model_history.history['loss'], label="loss")
plt.legend(['train'], loc= "upper right")
```

```python
X_test = np.array([300, 350]).reshape((1,1,2))
y_pred = model.predict(X_test, verbose=0)
```
