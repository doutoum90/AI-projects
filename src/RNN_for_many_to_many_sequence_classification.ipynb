```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import Input, Activation, Dense, Flatten, Dropout, LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.models import Model
```

```python
X1 = list()
X2 = list()

X = list()
y = list()

X1 = [x for x in range(6, 361,6)]
X2 = [x for x in range(7, 421, 7)]

y = [y for y in range(36, 391, 6)]
X = np.column_stack((X1, X2))
X = np.array(X).reshape(20, 3, 2)
y = np.array(y).reshape(20, 3, 1)

```

```python
def create_lstm_many_to_many_model():
    input_layer = Input(shape=(3,2))
    lstm1 = LSTM(100, activation="relu")(input_layer)
    rv = RepeatVector(3)(lstm1)
    lstm2 = LSTM(100, activation="relu", return_sequences=True)(rv)

    output_layer = TimeDistributed(Dense(1))(lstm2)
    model = Model(input_layer, output_layer)
    model.compile(loss="mse", optimizer='adam')
    return model
```

```python
from tensorflow.keras.utils import plot_model
model = create_lstm_many_to_many_model()
plot_model(model, to_file="../images/model_lstm_plot4.png", show_shapes=True, show_layer_names=True)
```

```python
model_history = model.fit(X, y, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt 
plt.plot(model_history.history['loss'], label="loss")
plt.legend(['train'], loc= "upper right")
```

```python
X1 = [600, 606, 612]
X2 = [700, 707, 714]

X_test = np.column_stack((X1, X2)).reshape((1,3,2))
y_pred = model.predict(X_test, verbose=0)
```
