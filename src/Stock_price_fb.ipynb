```python
import pandas as pd
import numpy as np
fb_complete_data = pd.read_csv("../data/FB.csv")
fb_complete_data.head()
```

```python
fb_train_processed = fb_complete_data[['Open']].values
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range= (0,1))
fb_train_scaled = scaler.fit_transform(fb_train_processed)
```

```python
len(fb_train_scaled)
```

```python
fb_training_features = []
fb_training_labels = []
for i in range(60,len(fb_train_scaled)):
    fb_training_features.append(fb_train_scaled[i-60:i, 0])
    fb_training_labels.append(fb_train_scaled[i,0])

X_train = np.array(fb_training_features)
y_train = np.array(fb_training_labels)

print(X_train.shape)
print(y_train.shape)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1], 1))
```

```python
import matplotlib.pyplot as plt 
from tensorflow.keras.layers import Input, Activation, Dense, Flatten, Dropout, LSTM
from tensorflow.keras.models import Model
```

```python
input_layer = Input(shape=(X_train.shape[1],1))
lstm1 = LSTM(100, activation="relu", return_sequences=True)(input_layer)
do1 = Dropout(0.2)(lstm1)
lstm2 = LSTM(100, activation="relu", return_sequences=True)(do1)
do2 = Dropout(0.2)(lstm2)
lstm3 = LSTM(100, activation="relu", return_sequences=True)(do2)
do3 = Dropout(0.2)(lstm3)
lstm4 = LSTM(100, activation="relu")(do3)
do4 = Dropout(0.2)(lstm4)
output_layer = Dense(1)(do4)
model = Model(input_layer, output_layer)
model.compile(loss="mse", optimizer='adam')
```

```python
print(X_train.shape, y_train.shape)
y_train = y_train.reshape(-1, 1)
print(X_train.shape, y_train.shape)
```

```python
model_history = model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```
